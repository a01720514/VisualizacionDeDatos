```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
df=pd.read_csv('datasets/antropometria-dataset-1.csv')
df.head()
```

## 1

```python
sns.displot(df['cintura'],
            bins=15)
```

## 2

```python
sns.lmplot(x='peso',
           y='cintura',data=df,hue='sexo')
```

## 3

```python
my_pivot=pd.pivot_table(df,
                        index='folio',
                        values='cintura',
                        columns='sexo'
                       )
my_pivot
```

```python
sns.boxplot(data=my_pivot.loc[:,[1]],orient='h')
```
