```python
sns.boxplot(data=df.loc[:,['edad', 'meses', 'peso','talla', 'emb', 'temb', 'cintura', 'cadera', 'sistol','diastol']])
```

```python
sns.boxplot(data=df.loc[df['sexo']==1, #cambair variable
                        ['edad', 'meses', 'peso','talla', 'emb', 'temb', 'cintura', 'cadera', 'sistol','diastol']]
           ).set_title('Box plot de Genero 1')
```

```python
sns.boxplot(data=df.loc[df['sexo']==2, #cambiar variable
                        ['edad', 'meses', 'peso','talla', 'emb', 'temb', 'cintura', 'cadera', 'sistol','diastol']]
           ).set_title('Box plot de Genero 2')
```

```python
dfN=df.loc[:,'peso'] #cambiar variable
dfN=dfN.transpose()
dfN.plot(kind='hist') #cambiar tipo de grafico
```

```python
dfN=df.loc[df['sexo']==1,'peso'] #cambiar variable
dfG=dfN.transpose()
dfG.plot(kind='hist').set_title('Histograma de Genero 1') #cambiar tipo de grafico
```

```python
dfN=df.loc[df['sexo']==2,'peso'] #cambiar variable
dfG=dfN.transpose()
dfG.plot(kind='hist').set_title('Histograma de Genero 2') #cambiar tipo de grafico
```

```python
sns.lmplot(x='peso',
           y='cintura',
           data=df,
           fit_reg=False,
           #hue ='sexo'
          )
plt.title('Diagrama de dispersión de peso vs cintura', fontsize=17)
plt.xticks(fontsize=15)
plt.show()
```
