## antropometria

```python
df=pd.read_csv('datasets/antropometria-dataset.csv')
df.head()
```

```python
my_pivot=pd.pivot_table(df,
                        index='desc_ent',
                        values='peso',
                        #aggfunc=['std','sum','mean','median','count']
                        aggfunc='mean',
                       )
my_pivot
```

```python
my_pivot=pd.pivot_table(df,
                        index='desc_ent',
                        values='peso',
                        #aggfunc=['std','sum','mean','median','count']
                        aggfunc='median',
                        columns='sexo',
                       )
my_pivot
```

```python
my_pivot=pd.pivot_table(df,
                        index='desc_ent',
                        values='peso',
                        #aggfunc=['std','sum','mean','median','count']
                        aggfunc='mean',
                        columns='sexo'
                       )
my_pivot
```

## adult

```python
df=pd.read_csv('datasets/adult.csv')
df.head()
```

```python
df.columns=list(map(str,df.columns))
```

```python
df.columns
```

```python
my_pivot=pd.pivot_table(df,
                        index=['marital-status','education'],
                        values=['hours-per-week'],
                        #aggfunc=['std','sum','mean','median','count']
                        aggfunc='mean',
                       )
my_pivot
```

```python
my_pivot.head(67)
```

```python
my_pivot=pd.pivot_table(df,
                        index=['education','marital-status'],
                        values=['age','hours-per-week'],
                        #aggfunc=['std','sum','mean','median','count']
                        aggfunc='median',
                        columns='sex',
                       )
my_pivot
```
