```python
import matplotlib.pyplot as plt 
import seaborn as sns
import statsmodels.api as sm 
import numpy as np
from sklearn import linear_model
```

```python
modelA= sm.OLS.from_formula('Salary~Age', data=df)
resultA= modelA.fit()
resultA.summary()
```

### Observaciones
* Variable dependiente: 'Salary'
* Variable independiente: 'Age'
* R-sq: 0.380; por lo tanto, este modelo no es tan bueno ya que solo explica en 38% de la avriación basada en la variable 'Age'
...
* Df model: 1
* Intercept... > 0.05, no es estadisticamente sigificativa para el modelo

` Salary =  349.4458 + 48.0932*Age`

```python
print('Salary of a person whose Age is 33: \t', 349.4458 + 48.0932*33)
df.iloc[:2,:]
```

```python
print('La diferencia es \t:', 1500-1936.5214)
print('La diferencia es \t:', 1200-1936.5214)
resultA.resid
```

```python
plt.figure(figsize=(15,10))
sns.set(font_scale=1.7)

reg_plot=sns.regplot(x='Age',
            y='Salary',
            data=df,
            marker='+',
            color='green',
           scatter_kws={'s':200},
           )
reg_plot.set_title('Regression plot of Salary vs. Age')
reg_plot.text(42,1500,'Salary =  349.4458 + 48.0932*Age',
             color='green',
             weight='semibold',
             size='medium')
```

```python
RESIA=resultA.resid
RESIA
```

```python
resi_plot=sns.scatterplot(x=resultA.fittedvalues,
                          y=RESIA)
```

## 2. Pregunta de investigación:

### Observaciones
* Variable dependiente: 'Salary'
* Variable independiente: 'Age' y 'Gender'
* R-sq: 0.408; por lo tanto, este modelo no es tan bueno ya que solo explica en 40.8% de la avriación basada en la variable 'Age'
...
* Df model: 2
* Intercept... > 0.05, no es estadisticamente sigificativa para el modelo

` Salary =  749.8157 + 34.2211*Age + 306.7169*Gender`

## 3. Pregunta de investigación:
¿Cuál es el impacto de Age, Gender, Academic_degree en Salary?

Salary = b0 + b1*Age + b2*Salary + b3*Academic_degree

```python
modelC= sm.OLS.from_formula('Salary~Age+Gender+Academic_degree', data=df)
resultC= modelC.fit()
resultC.summary()
```

### Observaciones
* Variable dependiente: 'Salary'
* Variable independiente: 'Age', 'Gender' y 'Academic_degree'
* R-sq: 0.483; por lo tanto, este modelo no es tan bueno ya que solo explica en 48.3% de la avriación basada en la variable 'Age'
...
* Df model: 4
* Intercept... > 0.05, no es estadisticamente sigificativa para el modelo

` Salary =  815.5485 + 36.3765*Age + 336.1028*Gender(Male) - 373.1962*Academic_degree(Master) - 43.8495*Academic_degree(No)`

### Case study 1:
DF Model: 4
* Age: 48
* Male = True
* Academic_degree_Master = True
* Academic_degree_No = False

```python
print('El salario de una persona con age 48, male y master \t:',
      815.5485 + 36.3765*48 + 336.1028*1 - 373.1962*1 - 43.8495*0)
df.iloc[-1:,:]
```

### Case study 2:
DF Model: 4
* Age: 33
* Male = False
* Academic_degree_Master = False
* Academic_degree_No = False

```python
print('El salario de una persona con age 33, female y bachelor \t:',
      815.5485 + 36.3765*33 + 336.1028*0 - 373.1962*0 - 43.8495*0)
df.iloc[:2,:]
```

```python
resultC.resid
```
