## Pregunta de investigaci√≥n:1
### ¬øExiste alguan correlaci√≥n entre 'Wingspan' y 'CWDistance'?

```python
sns.lmplot(x='Wingspan',
           y='CWDistance',
           data=df,
           fit_reg=False,
           hue ='Gender'
          )
plt.title('Scatter plot of Wingspan vs CWDistance', fontsize=17)
plt.xticks(fontsize=15)
plt.show()
```

### Conclusi√≥n
No se encontro ninguna correlacion entre 'Wingspan' y 'CWDistance'
